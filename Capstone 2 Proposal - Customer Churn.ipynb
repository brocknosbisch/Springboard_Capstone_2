#### What is the problem you want to solve? 
The problem I am trying to solve is pretty common in most utility companies.  What is the chance (score) that a person will turn on in the next XX days.  

#### Who is your client and why do they care about this problem? In other words, what will your client do or decide based on your analysis that they wouldn’t have done otherwise? 
My client is a natural gas company.  Being able to predict when a customer will turn on is crucial to the natural gas business. By knowing a range when a customer will turn on then a salesrep can reach out to the customer to get them to turn on early.  This is especially important when it initially gets cold in the fall and winter.  This surge puts a strain on employees since there is more work to be done than the number of employees.  Being able to get a customer to turn on early will reduce that strain and save the company money by reducing the amount of overtime worked.
