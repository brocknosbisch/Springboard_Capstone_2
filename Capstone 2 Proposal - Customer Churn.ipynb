#### What is the problem you want to solve? 
The overall problem I am trying to solve is pretty common in most consumer businesses, Customer Churn.  I will be going a few steps further than just looking at who is leaving though.  I want to:
* Predict if a customer will turn off.
* Give a probability of a customer turning off in a certain time frame (tbd).

#### Who is your client and why do they care about this problem? In other words, what will your client do or decide based on your analysis that they wouldn’t have done otherwise? 
My client is any energy company.  Being able to predict when a customer will turn off is crucial to the energy business.  Being in the natural gas industry means that several customers will turn off in the spring and turn on during the fall whenever it gets cold.  By knowing what customers are considered “Seasonal Customers”, customer reps can reach out to the customers to see how we can keep them on year-round.
Also, being able to predict when a customer will permanently turn-off (not turn back on for 1+ year) can help in marketing to that customer to try to keep them as a customer.<br>
